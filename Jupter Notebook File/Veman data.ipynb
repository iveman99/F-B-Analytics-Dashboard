```python
#Q1 Conversion rate by restaurant name
restaurant_conv = (
    df.groupby('REST_NAME')
      .agg(
          total_orders=('TOTAL_ORDERS','sum'),
          total_menu_opens=('MENU_SESSIONS','sum'),
      )
)

restaurant_conv['conversion_rate'] = restaurant_conv['total_orders'] / restaurant_conv['total_menu_opens']

restaurant_conv.sort_values('conversion_rate', ascending=False).head(10)
```

```python
#Q4 orders + GMV per restaurant per month:
monthly_restaurant = (
    df.groupby(['REST_NAME', 'month_year'])
      .agg(
          monthly_orders=('TOTAL_ORDERS', 'sum'),
          monthly_gmv=('GMV', 'sum')
      )
      .reset_index()
)
monthly_restaurant.head()

monthly_restaurant = (
    df.groupby(['REST_NAME', 'month_year'])
      .agg(
          monthly_orders=('TOTAL_ORDERS', 'sum'),
          monthly_gmv=('GMV', 'sum')
      )
      .reset_index()
)
monthly_restaurant.head()
```

```python
#Q4
monthly_restaurant = monthly_restaurant.sort_values(['REST_NAME', 'month_year'])

monthly_restaurant['orders_mom_growth'] = monthly_restaurant.groupby('REST_NAME')['monthly_orders'].pct_change()
monthly_restaurant['gmv_mom_growth'] = monthly_restaurant.groupby('REST_NAME')['monthly_gmv'].pct_change()

monthly_restaurant.head()
```

```python
growing_restaurants = monthly_restaurant[
    (monthly_restaurant['orders_mom_growth'] > 0) |
    (monthly_restaurant['gmv_mom_growth'] > 0)
]

growing_restaurants
```

```python
#Q5

# Base metrics per restaurant
restaurant_summary = (
    df.groupby('REST_NAME')
      .agg(
          total_orders=('TOTAL_ORDERS', 'sum'),
          total_menu_opens=('MENU_SESSIONS', 'sum'),
          total_gmv=('GMV', 'sum')
      )
      .reset_index()
)

# Add conversion
restaurant_summary['conversion_rate'] = restaurant_summary['total_orders'] / restaurant_summary['total_menu_opens']

restaurant_summary
```

```python
last_month_growth = (
    monthly_restaurant
    .groupby('REST_NAME')
    .apply(
        lambda x: x.sort_values('month_year').iloc[-1][['orders_mom_growth','gmv_mom_growth']],
        include_groups=False
    )
    .reset_index()
)
last_month_growth 
```

```python
promo_df = restaurant_summary.merge(last_month_growth, on='REST_NAME', how='left')

promo_df
```

```python
promo_df['promo_priority'] = promo_df.apply(
    lambda row:
        'High Traffic, Low Conversion' if (row['total_menu_opens'] > promo_df['total_menu_opens'].median() 
                                           and row['conversion_rate'] < promo_df['conversion_rate'].median())
        else 'Low Traffic, High Conversion' if (row['total_menu_opens'] < promo_df['total_menu_opens'].median() 
                                               and row['conversion_rate'] > promo_df['conversion_rate'].median())
        else 'Growing MoM' if (row['orders_mom_growth'] > 0 and row['gmv_mom_growth'] > 0)
        else 'Low Priority',
    axis=1
)

promo_df
```
